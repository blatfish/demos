```python
## housekeeping
import numpy as np
import matplotlib.pyplot as plt
```

```python
## function to difference a series to given order
def difference(x, order):
    n = len(x) # length of input
    working = np.copy(x) # initialize output at order 0
    for i in range(order): # to order
        working -= np.roll(working, 1) # difference
    return working[order:len(working)]
```

```python
## test output
n = 50 # length of input
x = np.arange(n) # independent variable
y =  x**2 + np.random.normal(scale=5, size=n) # differentiable function + noise
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 6))
ax1.scatter(x, y, marker=".") # plot original
ax1.title.set_text("order 0 (undifferentiated)")
ax2.scatter(x[1:], difference(y, 1), marker=".") # plot first difference
ax2.title.set_text("order 1 (first derivative)")
ax3.scatter(x[2:], difference(y, 2), marker=".") # plot second difference
ax3.title.set_text("order 2 (second derivative)")
plt.show()
```
